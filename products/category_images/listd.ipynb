```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

# Configuration
excel_file = 'db.xlsx'
sheet_name = 'Categories'  # Name of the sheet in the Excel file
table_name = 'products_category'  # Name of the SQL table
```

```python

# Database connection details
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'root',
    'database': 'sclgaming_site'
}
```

```python

def insert_excel_to_sql(excel_file, sheet_name, table_name, db_config):
    try:
        # Read data from Excel file
        df = pd.read_excel(excel_file, sheet_name=sheet_name)

        # Establish database connection
        connection = mysql.connector.connect(**db_config)
        cursor = connection.cursor()

        # Create SQL INSERT statement
        columns = ', '.join(df.columns)
        values_placeholder = ', '.join(['%s'] * len(df.columns))
        sql = f"INSERT INTO {table_name} ({columns}) VALUES ({values_placeholder})"

        # Insert data into SQL table
        for index, row in df.iterrows():
            cursor.execute(sql, tuple(row))

        # Commit changes
        connection.commit()
        connection.close()
        print(f"Data inserted into {table_name} successfully.")
        

    except Error as e:
        print(f"Error: {e}")
```
